## Add() Function

```python
def add(v: Vector, w:Vector) -> Vector:
    """Adds corresponding elements"""
    assert len(v) == len(w), "vectors must be the same length"
    return [v_i+w_i for v_i, w_i in zip(v,w)]
```

### Function Breakdown

- **`v: Vector` and `w: Vector`**: These are the function's parameters. The `Vector` type hint suggests that both `v` and `w` should be of a type that represents a vector, which is typically a list or array of numbers.
- **`-> Vector`**: This indicates that the function is expected to return a `Vector` (again, typically a list or array of numbers).

- **`assert`**: Used to enforce a condition. If the condition is `False`, it raises an `AssertionError` with the provided message.
- **`len(v) == len(w)`**: Checks if the lengths of vectors `v` and `w` are the same.
- **`"vectors must be the same length"`**: Error message displayed if the vectors' lengths are not equal. This assertion ensures that the vectors are of the same length before attempting to add them element-wise.


### Example Usage of add() function

```python
v = [1,2,3]
w = [4,5,6]
print(add(v,w))
assert add(v,w) == [5,7,9]
```

```python
import matplotlib.pyplot as plt

v = [1,2]
w = [2,1]
result = add(v, w)

plt.figure()

# Plot vector v from (0, 0) to (1, 2)
plt.plot([0, v[0]], [0, v[1]], color='r', marker='o', label='v (1,2)')

# Plot vector w from the tip of vector v
plt.plot([v[0], result[0]], [v[1], result[1]], color='b', marker='o', label='w (2,1)')

# Plot the resultant vector from (0, 0) to (3, 3)
plt.plot([0, result[0]], [0, result[1]], color='g', linestyle='--', marker='o', label='v + w (3,3)')

# Adding arrows to the vectors
plt.annotate('', xy=(v[0], v[1]), xytext=(0, 0), arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate('', xy=(result[0], result[1]), xytext=(v[0], v[1]), arrowprops=dict(facecolor='blue', shrink=0.05))
plt.annotate('', xy=(result[0], result[1]), xytext=(0, 0), arrowprops=dict(facecolor='green', shrink=0.05, linestyle='--'))

# Set the x and y axis limits
plt.xlim(-0.5, 4)
plt.ylim(-1, 4)

# Add grid, labels, and title
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Addition: Tail-to-Tip Visualization with Line Plots')
plt.legend()

plt.show()
```

### Why Vectors Start from (0, 0)

In vector mathematics, especially when visualizing vector operations like addition, vectors are typically represented as directed arrows starting from the origin `(0, 0)` unless otherwise specified. Heres why:

#### Standard Representation
- The origin `(0, 0)` is the default starting point for vectors in 2D space when not considering their position in the context of a coordinate system. This makes it easier to visualize the direction and magnitude of the vector without worrying about its position.

#### Vector Addition
- When visualizing vector addition, the first vector is typically drawn from the origin `(0, 0)` to its endpoint. The second vector is then drawn starting from the tip of the first vector to its endpoint.
- The resultant vector (which is the sum) is usually drawn from the origin `(0, 0)` directly to the endpoint of the second vector.

#### Generalization
- Starting vectors from the origin simplifies the general understanding of vector operations, such as addition, subtraction, and scalar multiplication, because it eliminates the need to account for any initial displacement. The vectors are treated as "free vectors," meaning their only defining characteristics are their direction and magnitude, not their specific position.

#### Consistency in Visualization
- By starting at `(0, 0)`, the visualization is consistent with how vector addition is typically taught and understood in mathematics. This consistency helps in recognizing patterns and understanding the effects of vector operations.

#### Example in the Original Code
- **Red Arrow (`v`)**: The vector `v = [1, 2]` starts at `(0, 0)` and ends at `(1, 2)`. This shows the direction and magnitude of `v`.
- **Blue Arrow (`w`)**: The vector `w = [2, 1]` starts at `(1, 2)` (the tip of `v`) and ends at `(3, 3)`.
- **Green Arrow (Resultant Vector)**: The resultant vector, representing `v + w`, starts at `(0, 0)` and ends at `(3, 3)`.

This approach of starting from `(0, 0)` is helpful because it allows you to see how vectors combine and what the final resultant vector looks like in a simple and straightforward way.

## Subtract() function

```python
def subtract(v:Vector,w:Vector) -> Vector:
    assert len(v)==len(w), "vectors must be the same length"
    return [v_i-w_i for v_i,w_i in zip(v,w)]
```

### Example Usage of subtract() function

```python
v = [5,7,9]
w = [4,5,6] 
print(subtract(v,w))
assert subtract(v,w) == [1,2,3]
```

```python
v = [3,2]
w = [2,1]
result = subtract(v, w)

plt.figure()

plt.plot([0, v[0]], [0, v[1]], color='r', marker='o', label='v (1,2)')
plt.plot([v[0], result[0]], [v[1], result[1]], color='b', marker='o', label='w (2,1)')
plt.plot([0, result[0]], [0, result[1]], color='g', linestyle='--', marker='o', label='v + w (3,3)')


plt.annotate('', xy=(v[0], v[1]), xytext=(0, 0), arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate('', xy=(result[0], result[1]), xytext=(v[0], v[1]), arrowprops=dict(facecolor='blue', shrink=0.05))
plt.annotate('', xy=(result[0], result[1]), xytext=(0, 0), arrowprops=dict(facecolor='green', shrink=0.05, linestyle='--'))


plt.xlim(-0.5, 4)
plt.ylim(-1, 4)


plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Subtraction: Tail-to-Tip Visualization with Line Plots')
plt.legend()

plt.show()
```

# Sum of vector list

A new vectors that has first component as a sum of all components, and second components is a sum of all second components, and so on.

```python
def vector_sum(vectors: List[Vector]) -> Vector:
    #Check the vectors is not empty
    assert vectors, "no vectors provided!"

    #Chheck the vectors are all the same size
    num_elements = len(vectors[0])
    assert all(len(v) == num_elements for v in vectors), "different sizes!"

    return [sum(vector[i] for vector in vectors)
            for i in range(num_elements)]

assert vector_sum([[1,2],[3,4],[5,6],[7,8]]) == [16,20]
```
